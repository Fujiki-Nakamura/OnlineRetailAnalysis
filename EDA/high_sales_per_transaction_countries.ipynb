```python
import time 

from IPython.core.display import display
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
import numpy as np
import seaborn as sns
sns.set(font='Osaka')
import pandas as pd
```

```python
t0 = time.time()
data_dir = '../data/'
file = 'df_sales.csv'
df_sales = pd.read_csv(data_dir + file)
print('Load in ', time.time() - t0)
```

```python
df_sales.head()
```

```python
# Netherlands
# 各 Customer　の売上構成比
country = 'Netherlands'

df_sales_by_country_customer = pd.DataFrame(df_sales.groupby(['Country', 'CustomerID'])['Sales'].sum()).reset_index()
df_the_country = df_sales_by_country_customer[df_sales_by_country_customer['Country'] == country]
df_the_country['Sales_%'] = \
(df_the_country['Sales'] / df_the_country['Sales'].sum()) * 100
df_the_country = df_the_country.sort_values('CustomerID')

# 各 Customer のトランザクション数とその構成比
country = 'Netherlands'

df_transactions_country_customer = \
    pd.DataFrame(df_sales.groupby(['Country', 'CustomerID']).size().reset_index())

df_the_country['Transaction'] = \
df_transactions_country_customer[df_transactions_country_customer['Country'] == country]\
.sort_values('CustomerID').rename(columns={0: 'Transaction'})['Transaction']
df_the_country['Sales/Transaction'] = \
df_the_country['Sales'] / df_the_country['Transaction']

df_the_country.sort_values('Sales', ascending=False)
```

```python
def make_df_the_country(df_sales, country):
    # 各 Customer　の売上構成比
    df_sales_by_country_customer = pd.DataFrame(df_sales.groupby(['Country', 'CustomerID'])['Sales'].sum()).reset_index()
    df_the_country = df_sales_by_country_customer[df_sales_by_country_customer['Country'] == country]
    df_the_country['Sales_%'] = \
    (df_the_country['Sales'] / df_the_country['Sales'].sum()) * 100
    df_the_country = df_the_country.sort_values('CustomerID')

    # 各 Customer のトランザクション数とその構成比
    df_transactions_country_customer = \
        pd.DataFrame(df_sales.groupby(['Country', 'CustomerID']).size().reset_index())

    df_the_country['Transaction'] = \
    df_transactions_country_customer[df_transactions_country_customer['Country'] == country]\
    .sort_values('CustomerID').rename(columns={0: 'Transaction'})['Transaction']
    df_the_country['Sales/Transaction'] = \
    df_the_country['Sales'] / df_the_country['Transaction']

    return df_the_country

country = 'Australia'
df_the_country = make_df_the_country(df_sales, country)
df_the_country.sort_values('Sales', ascending=False)
```

```python
# NOTE:
# Japan は 売上1位が Sales/Transaction も1位であるわけではない。

country = 'Japan'
make_df_the_country(df_sales, country).sort_values('Sales', ascending=False)
```
